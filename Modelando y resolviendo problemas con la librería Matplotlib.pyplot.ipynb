```python
import matplotlib.pyplot as plt
import numpy as np

# Crear datos de la onda senoidal
x = np.linspace(0, 2 * np.pi, 100)
y = np.sin(x)

# Crear la figura y el eje
fig, ax = plt.subplots()

# Dibujar la onda senoidal inicial
line, = ax.plot(x, y, label='Senoidal')

# Configurar leyenda y título
ax.legend()
ax.set_title('Onda Senoidal')

# Función para actualizar la onda senoidal con una frecuencia variable
def update_sine_wave(frequency):
    new_y = np.sin(frequency * x)
    line.set_ydata(new_y)
    ax.set_title(f'Onda Senoidal - Frecuencia: {frequency}')
    plt.draw_if_interactive()

# Mostrar el gráfico interactivo
plt.ion()

# Actualizar la onda senoidal con frecuencias diferentes de forma interactiva
for freq in [1, 2, 3]:
    update_sine_wave(freq)
    plt.pause(1)  # Pausa para ver la actualización antes de pasar a la siguiente frecuencia

# Desactivar el modo interactivo al finalizar
plt.ioff()

# Mostrar el gráfico final
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Crear datos de la onda senoidal
x = np.linspace(0, 2 * np.pi, 100)
frequencies = [1, 2, 3]  # Lista de frecuencias

# Crear la figura y el eje
fig, ax = plt.subplots()

# Dibujar la onda senoidal inicial
line, = ax.plot(x, np.sin(x * frequencies[0]), label=f'Senoidal - Frecuencia: {frequencies[0]:.2f}')

# Configurar leyenda y título
ax.legend()
ax.set_title('Onda Senoidal')

# Función para actualizar la onda senoidal con diferentes frecuencias
def update_sine_wave(frequency):
    line.set_ydata(np.sin(x * frequency))
    ax.set_title(f'Onda Senoidal - Frecuencia: {frequency:.2f}')
    plt.draw_if_interactive()

# Mostrar el gráfico interactivo
plt.ion()

# Actualizar la onda senoidal con frecuencias diferentes en intervalos de un segundo
for freq in frequencies:
    update_sine_wave(freq)
    plt.pause(1)  # Pausa de un segundo

# Desactivar el modo interactivo al finalizar
plt.ioff()

# Mostrar el gráfico final
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Crear datos de la onda senoidal
x = np.linspace(0, 2 * np.pi, 100)
frequencies = [1, 2, 3]  # Frecuencias para las ondas senoidales
titles = [f'Onda Senoidal - Frecuencia: {freq}' for freq in frequencies]

# Crear la figura y el eje
fig, ax = plt.subplots()

# Configurar título
ax.set_title(titles[0])

# Función para dibujar una onda senoidal y actualizar el título
def draw_sine_wave(frequency):
    y = np.sin(frequency * x)
    ax.clear()  # Limpiar el eje antes de dibujar una nueva onda
    ax.plot(x, y, label='Senoidal')
    ax.legend()

# Dibujar las ondas senoidales en intervalos de tiempo de un segundo
for i, freq in enumerate(frequencies):
    draw_sine_wave(freq)
    plt.pause(1)
    if i < len(frequencies) - 1:
        ax.set_title(titles[i+1])

# Mostrar el gráfico final
plt.show()
```

```python
import matplotlib.pyplot as plt
import pandas as pd

paises_sudamerica = ["Argentina", "Bolivia", "Brasil", "Chile", "Colombia",
                    "Ecuador", "Paraguay", "Perú", "Uruguay", "Venezuela"]

copas_mundial = {
    "Argentina": 2,
    "Brasil": 5,
    "Uruguay": 2,
    "Paraguay": 0,
    "Chile": 0,
    "Colombia": 0,
    "Ecuador": 0,
    "Perú": 0,
    "Bolivia": 0,
    "Venezuela": 0
}

# Gráfico 1: Barras horizontales
plt.subplot(131)
plt.barh(paises_sudamerica, copas_mundial.values(), color="green")
plt.xlabel("Cantidad de Copas del Mundo")
plt.ylabel("Países")
plt.title("Países sudamericanos ganadores")

# Gráfico 2: Gráfico circular
plt.subplot(132)
copas_series = pd.Series(copas_mundial)
copas_series.replace(to_replace=0, method="ffill", inplace=True)
plt.pie(copas_series, labels=copas_series.index, autopct="%1.1f%%")
plt.title("Porcentaje de Copas del Mundo")

# Gráfico 3: Barras apiladas
plt.subplot(133)
ganadores = copas_series[copas_series > 0].index.to_list()
no_ganadores = copas_series[copas_series == 0].index.to_list()
df_barras = pd.DataFrame({
    "Ganadores": [copas_mundial[pais] for pais in ganadores],
    "No Ganadores": [0 for _ in range(len(ganadores))]
}, index=ganadores)
df_barras.plot.bar(stacked=True)
plt.xlabel("Países")
plt.ylabel("Cantidad de Copas del Mundo")
plt.title("Comparación de Copas del Mundo")

plt.show()
```

```python
import matplotlib.pyplot as plt

# Lista de todos los países de Sudamérica
paises_sudamerica = ['Argentina', 'Bolivia', 'Brasil', 'Chile', 'Colombia', 'Ecuador', 'Guyana', 'Paraguay', 'Perú', 'Surinam', 'Uruguay', 'Venezuela']

# Número de Copas del Mundo ganadas por cada país
copas_mundo = {'Argentina': 3, 'Brasil': 5, 'Uruguay': 2}

# Asegurarse de que todos los países estén en el diccionario, si no han ganado ninguna copa, se pone 0
for pais in paises_sudamerica:
    if pais not in copas_mundo:
        copas_mundo[pais] = 0

# Crear listas separadas para los nombres de los países y el número de copas
paises = list(copas_mundo.keys())
copas = list(copas_mundo.values())

# Crear el primer gráfico (barras)
plt.figure(figsize=(10, 5))
plt.bar(paises, copas, color='green')
plt.xlabel('Países')
plt.ylabel('Número de Copas del Mundo')
plt.title('Número de Copas del Mundo ganadas por países de Sudamérica')
plt.xticks(rotation=90)
plt.show()

# Crear el segundo gráfico (pastel)
plt.figure(figsize=(10, 5))
plt.pie(copas, labels=paises, autopct='%1.1f%%')
plt.title('Porcentaje de Copas del Mundo ganadas por países de Sudamérica')
plt.show()

# Crear el tercer gráfico (lineas)
plt.figure(figsize=(10, 5))
plt.plot(paises, copas, marker='o')
plt.xlabel('Países')
plt.ylabel('Número de Copas del Mundo')
plt.title('Número de Copas del Mundo ganadas por países de Sudamérica')
plt.xticks(rotation=90)
plt.grid(True)
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import requests

# Obtener datos de tasa de desempleo por estado desde API
api_url = 'https://api.datos.gov/v1/unemployment' 
response = requests.get(api_url)
data = response.json()

# Convertir a DataFrame de Pandas
df = pd.DataFrame(data) 

# Crear arrays de latitud, longitud y tasa de desempleo
lats = df['lat'].values  
lons = df['lon'].values
rates = df['rate'].values

# Generar grilla para interpolación
xi, yi = np.linspace(min(lons), max(lons), 100), np.linspace(min(lats), max(lats), 100)
xi, yi = np.meshgrid(xi, yi)

# Interpolación para crear superficie contínua
interp = scipy.interpolate.griddata((lons, lats), rates, (xi, yi), method='linear') 

# Graficar mapa de contorno    
plt.contourf(xi, yi, interp, 10, cmap='RdGy')
plt.colorbar()
plt.title('Tasa de Desempleo por Estado')
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Generar datos de ejemplo
x = np.linspace(-5, 5, 100)
y = np.linspace(-5, 5, 100)
X, Y = np.meshgrid(x, y)
Z = np.sin(np.sqrt(X**2 + Y**2))

# Crear un mapa de contorno
plt.contourf(X, Y, Z, 20, cmap='RdGy')
plt.colorbar()  # Barra de color que muestra la escala de colores
plt.title('Mapa de Contorno')
plt.xlabel('Eje X')
plt.ylabel('Eje Y')
plt.show()
```

```python
from collections import Counter
import matplotlib.pyplot as plt

# Suponiendo que 'pre_review' contiene texto, podemos tokenizar y contar
words = preprocessed_df['pre_review'].str.split(expand=True).stack()
word_counts = Counter(words)

# Seleccionar las 10 palabras más comunes para visualizar
most_common_words = word_counts.most_common(10)
words, frequencies = zip(*most_common_words)

plt.bar(words, frequencies)
plt.xticks(rotation=45)
plt.xlabel('Palabras')
plt.ylabel('Frecuencia')
plt.title('Top 10 Palabras más Frecuentes')
plt.show()
```
