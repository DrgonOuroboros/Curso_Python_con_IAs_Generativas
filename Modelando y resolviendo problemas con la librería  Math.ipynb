```python
import math

# Cálculo de e^2 - 1 usando math.expm1
resultado_expm1 = math.expm1(2)

# Cálculo de e^2 - 1 usando math.exp y resta
resultado_exp = math.exp(2) - 1

# Impresión de resultados
print("Resultado con math.expm1:", resultado_expm1)
print("Resultado con math.exp:", resultado_exp)

# Comparación de resultados
print("¿Los resultados son iguales?", resultado_expm1 == resultado_exp)
```

### Funciones hiperbólicas

```python
import math

# Definir la función tangente hiperbólica
def tanh(x):
    return (math.exp(x) - math.exp(-x)) / (math.exp(x) + math.exp(-x))

# Calcular tanh(2)
resultado = tanh(2)

# Imprimir el resultado
print(f"tanh(2) = {resultado}")
```

```python
import math
import numpy as np
import matplotlib.pyplot as plt

# Función seno hiperbólico 
def senh(x):
  return (math.e**x - math.e**(-x))/2

# Vector de puntos x  
x = np.arange(-5, 5, 0.1)  

# Vector de puntos y 
y = [senh(i) for i in x]

# Graficar
plt.plot(x,y)
plt.title("Función Seno Hiperbólico")
plt.xlabel("X")
plt.ylabel("f(x)")
plt.grid() 
plt.show()
```

```python
from scipy.special import erf
from scipy.stats import norm

def phi(x, tipo="cdf"):
  """
  Función que calcula la función de error (erf) o la CDF de la distribución normal estándar,
  dependiendo del valor del argumento "tipo".

  Parámetros:
    x: El valor para el que se calcula la función.
    tipo: Indica qué función se desea calcular: "erf" o "cdf".

  Retorno:
    El valor de la función erf o la CDF de la normal estándar para el valor x.
  """
  if tipo == "erf":
    return erf(x / sqrt(2.0))
  elif tipo == "cdf":
    return norm(0, 1).cdf(x)
  else:
    raise ValueError(f"Tipo de función no válido: {tipo}")

# Ejemplos de uso
x = 1.0
resultado_erf = phi(x, tipo="erf")
print(f"Función de error para x={x}: {resultado_erf}")

x = 0.5
resultado_cdf = phi(x, tipo="cdf")
print(f"CDF para x={x}: {resultado_cdf}")
```

```python
from math import sqrt, erfc

def erf(x):
  """
  Calcula la función de error (erf) para un valor x usando la función erfc.

  Parámetros:
    x: El valor para el que se calcula la función erf.

  Retorno:
    El valor de la función erf para el valor x.
  """
  return 1.0 - erfc(x / sqrt(2.0))

# Ejemplo de uso
x = 1.0
resultado_erf = erf(x)
print(f"Función de error para x={x}: {resultado_erf}")
```
